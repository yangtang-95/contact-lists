```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
import time

# Initialize the WebDriver
driver = webdriver.Chrome()

# Log into LinkedIn
driver.get('https://www.linkedin.com/login')
time.sleep(2)
username = driver.find_element_by_name('session_key')
username.send_keys('your_email')
password = driver.find_element_by_name('session_password')
password.send_keys('your_password')
password.send_keys(Keys.RETURN)
time.sleep(2)

# List of names and companies
contacts = [
    {"name": "John Doe", "company": "ABC Corp"},
    {"name": "Jane Smith", "company": "XYZ Inc"}
]

# Function to send connection requests
def send_connection_request(name, company):
    search_query = f"{name} {company} LinkedIn"
    driver.get('https://www.google.com')
    search_box = driver.find_element_by_name('q')
    search_box.send_keys(search_query)
    search_box.send_keys(Keys.RETURN)
    time.sleep(2)
    
    # Click on the first LinkedIn profile link from search results
    profile_link = driver.find_element_by_partial_link_text('LinkedIn')
    profile_link.click()
    time.sleep(2)
    
    # Click the connect button
    connect_button = driver.find_element_by_xpath("//button[contains(., 'Connect')]")
    connect_button.click()
    time.sleep(2)
    
    # Add a note and send the connection request
    add_note_button = driver.find_element_by_xpath("//button[contains(., 'Add a note')]")
    add_note_button.click()
    time.sleep(2)
    
    message_box = driver.find_element_by_name('message')
    message_box.send_keys(f"Hi {name}, I'd like to connect with you on LinkedIn.")
    time.sleep(2)
    
    send_button = driver.find_element_by_xpath("//button[contains(., 'Send')]")
    send_button.click()
    time.sleep(2)

# Loop through contacts and send connection requests
for contact in contacts:
    send_connection_request(contact['name'], contact['company'])

driver.quit()
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.options import Options
import time
import pandas as pd
from bs4 import BeautifulSoup

# Your LinkedIn credentials
email = "polakowa@rogers.com"
password = "Timisoara#13"

# Initialize the Chrome driver with options
chrome_options = Options()
#chrome_options.add_argument("--headless")
chrome_options.add_argument("--no-sandbox")
chrome_options.add_argument("--disable-dev-shm-usage")
driver = webdriver.Chrome(options=chrome_options)

# Go to the LinkedIn login page
driver.get("https://www.linkedin.com/login")
time.sleep(2)

# Enter your LinkedIn credentials
username = driver.find_element(By.ID, "username")
password_field = driver.find_element(By.ID, "password")
username.send_keys(email)
password_field.send_keys(password)

# Click the login button
login_button = driver.find_element(By.XPATH, '//*[@type="submit"]')
login_button.click()
time.sleep(5)

# Go to the company page (replace with your target company page URL)
driver.get("https://www.linkedin.com/company/building-knowledge-canada-inc./posts")
time.sleep(5)

# Scroll down to load more posts
for _ in range(5):  # Adjust the range to load more posts
    driver.find_element(By.TAG_NAME, 'body').send_keys(Keys.END)
    time.sleep(2)

# Click all "see more" buttons to reveal hidden parts of the posts
see_more_buttons = driver.find_elements(By.XPATH, '//button[contains(@class, "see-more")]')
for button in see_more_buttons:
    try:
        driver.execute_script("arguments[0].click();", button)
        time.sleep(1)  # Wait for the content to load
    except:
        continue  # Skip if the button is not clickable

# Extract page source and parse with BeautifulSoup
soup = BeautifulSoup(driver.page_source, 'html.parser')
posts = soup.find_all('div', class_='update-components-text relative update-components-update-v2__commentary')

# Extract text from each post
post_texts = []
for post in posts:
    text = post.get_text(separator=" ", strip=True)
    post_texts.append(text)

# Print the number of posts found
print(f"Number of posts found: {len(posts)}")
print(f"Number of post texts extracted: {len(post_texts)}")

# Save the post texts to an Excel file
df = pd.DataFrame(post_texts, columns=['Post Text'])
df.to_excel('linkedin_posts.xlsx', index=False)

# Close the driver
driver.quit()
```
